````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# ============================================================================
# UTILITY FUNCTIONS
# ============================================================================

def random_unit_quaternion():
    """Generate random unit quaternion (uniform on S³)"""
    q = np.random.randn(4)
    return q / np.linalg.norm(q)

def random_su2_rotation():
    """Generate random SU(2) rotation matrix"""
    theta = np.random.uniform(0, 2*np.pi)
    phi = np.random.uniform(0, np.pi)
    U = np.array([
        [np.cos(theta/2), -np.exp(-1j*phi)*np.sin(theta/2)],
        [np.exp(1j*phi)*np.sin(theta/2), np.cos(theta/2)]
    ])
    return U

def random_so_rotation(n):
    """Generate random SO(n) rotation matrix using QR decomposition"""
    A = np.random.randn(n, n)
    Q, R = np.linalg.qr(A)
    d = np.diag(np.sign(np.diag(R)))
    return Q @ d

def hopf_map(q):
    """Map S³ (quaternion) to S² (Bloch vector)"""
    a, b, c, d = q
    x = 2*(a*c + b*d)
    y = 2*(b*c - a*d)
    z = a**2 + b**2 - c**2 - d**2
    bloch = np.array([x, y, z])
    return bloch / np.linalg.norm(bloch)

def rotate_bloch_via_su2(bloch, U):
    """Rotate Bloch vector via SU(2) action"""
    # Convert Bloch to quaternion, apply U, convert back
    a, b, c, d = quaternion_from_bloch(bloch)
    q = np.array([a, b, c, d], dtype=complex)

    alpha = a + 1j*b
    beta = c + 1j*d

    alpha_rot = U[0,0]*alpha + U[0,1]*(-np.conj(beta))
    beta_rot = U[1,0]*alpha + U[1,1]*(-np.conj(beta))

    a_rot = np.real(alpha_rot)
    b_rot = np.imag(alpha_rot)
    c_rot = np.real(beta_rot)
    d_rot = np.imag(beta_rot)

    bloch_rot = hopf_map(np.array([a_rot, b_rot, c_rot, d_rot]))
    return bloch_rot

def quaternion_from_bloch(bloch):
    """Inverse of Hopf map: convert Bloch vector back to quaternion"""
    x, y, z = bloch
    a = np.sqrt((1 + z) / 2)
    b = 0
    c = x / (2*a + 1e-10)
    d = y / (2*a + 1e-10)
    q = np.array([a, b, c, d])
    return q / np.linalg.norm(q)

def geodesic_distance_s2(bloch1, bloch2):
    """Geodesic distance on S² (great circle distance)"""
    dot_prod = np.clip(np.dot(bloch1, bloch2), -1.0, 1.0)
    return np.arccos(dot_prod)

def geodesic_distance_sn(point1, point2):
    """Geodesic distance on general Sⁿ"""
    dot_prod = np.clip(np.dot(point1, point2), -1.0, 1.0)
    return np.arccos(dot_prod)

# ============================================================================
# ZETA VALUE COMPUTATIONS
# ============================================================================

def zeta_2():
    """ζ(2) = π²/6"""
    return np.pi**2 / 6

def zeta_4():
    """ζ(4) = π⁴/90"""
    return np.pi**4 / 90

def zeta_6():
    """ζ(6) = π⁶/945"""
    return np.pi**6 / 945

# ============================================================================
# S² CASE: Test ζ(2) with √3 factor
# ============================================================================

def test_s2_geodesics(trials=50000):
    """
    Generate pure states on S², apply random SU(2) rotations,
    measure geodesic distances, estimate ζ(2)
    """
    distances_sq = []

    for _ in range(trials):
        # Generate pure state (random quaternion → Bloch vector)
        q = random_unit_quaternion()
        bloch = hopf_map(q)

        # Apply random SU(2) rotation
        U = random_su2_rotation()
        bloch_rot = rotate_bloch_via_su2(bloch, U)

        # Compute geodesic distance
        dist = geodesic_distance_s2(bloch, bloch_rot)
        distances_sq.append(dist**2)

    distances_sq = np.array(distances_sq)

    mean_dist_sq = np.mean(distances_sq)
    std_error = np.std(distances_sq) / np.sqrt(trials)

    # Extract ζ(2) from E[distance²] = (4/3) × √3 × ζ(2) / √3
    # So: ζ(2) = (3/4) × E[distance²]
    zeta2_measured = (3/4) * mean_dist_sq

    return zeta2_measured, std_error, mean_dist_sq

# ============================================================================
# S⁴ CASE: Test for ζ(4) pattern with √5 factor
# ============================================================================

def hopf_map_s4_to_s3(point_s4):
    """Map S⁴ to S³ via generalized Hopf fibration"""
    # For a 5D point normalized to S⁴, take first 4 components as S³
    return point_s4[:4] / np.linalg.norm(point_s4[:4])

def test_s4_geodesics(trials=50000):
    """
    Generate points on S⁴, apply random SO(5) rotations,
    measure geodesic distances on S⁴
    """
    distances_sq = []

    for _ in range(trials):
        # Generate random point on S⁴
        point_s4 = np.random.randn(5)
        point_s4 = point_s4 / np.linalg.norm(point_s4)

        # Apply random SO(5) rotation
        rotation = random_so_rotation(5)
        point_s4_rot = rotation @ point_s4

        # Compute geodesic distance on S⁴
        dist = geodesic_distance_sn(point_s4, point_s4_rot)
        distances_sq.append(dist**2)

    distances_sq = np.array(distances_sq)

    mean_dist_sq = np.mean(distances_sq)
    std_error = np.std(distances_sq) / np.sqrt(trials)

    # If pattern holds: E[distance²] = (4/3) × √5 × ζ(4) / √5
    # So: ζ(4) = (3/4) × E[distance²]
    zeta4_measured = (3/4) * mean_dist_sq

    return zeta4_measured, std_error, mean_dist_sq

# ============================================================================
# S⁶ CASE: Test for ζ(6) pattern with √7 factor
# ============================================================================

def test_s6_geodesics(trials=30000):
    """
    Generate points on S⁶, apply random SO(7) rotations,
    measure geodesic distances on S⁶
    """
    distances_sq = []

    for _ in range(trials):
        # Generate random point on S⁶
        point_s6 = np.random.randn(7)
        point_s6 = point_s6 / np.linalg.norm(point_s6)

        # Apply random SO(7) rotation
        rotation = random_so_rotation(7)
        point_s6_rot = rotation @ point_s6

        # Compute geodesic distance on S⁶
        dist = geodesic_distance_sn(point_s6, point_s6_rot)
        distances_sq.append(dist**2)

    distances_sq = np.array(distances_sq)

    mean_dist_sq = np.mean(distances_sq)
    std_error = np.std(distances_sq) / np.sqrt(trials)

    zeta6_measured = (3/4) * mean_dist_sq

    return zeta6_measured, std_error, mean_dist_sq

# ============================================================================
# MAIN EXECUTION
# ============================================================================

print("=" * 80)
print("DIMENSIONAL ZETA FUNCTION HYPOTHESIS TEST")
print("=" * 80)
print()

# Test S²
print("Testing S² (3D ambient space)...")
z2_meas, z2_err, d2_sq = test_s2_geodesics(trials=50000)
z2_theory = zeta_2()
z2_corrected_theory = np.sqrt(3) * z2_theory

print(f"  E[distance²] = {d2_sq:.6f}")
print(f"  ζ(2) measured = {z2_meas:.6f} ± {z2_err:.6f}")
print(f"  ζ(2) theory = {z2_theory:.6f}")
print(f"  √3 × ζ(2) = {z2_corrected_theory:.6f}")
error_z2_pct = 100 * abs(z2_meas - z2_corrected_theory) / z2_corrected_theory
print(f"  Error: {error_z2_pct:.2f}%")
print()

# Test S⁴
print("Testing S⁴ (5D ambient space)...")
z4_meas, z4_err, d4_sq = test_s4_geodesics(trials=50000)
z4_theory = zeta_4()
z4_corrected_theory = np.sqrt(5) * z4_theory

print(f"  E[distance²] = {d4_sq:.6f}")
print(f"  ζ(4) measured = {z4_meas:.6f} ± {z4_err:.6f}")
print(f"  ζ(4) theory = {z4_theory:.6f}")
print(f"  √5 × ζ(4) = {z4_corrected_theory:.6f}")
error_z4_pct = 100 * abs(z4_meas - z4_corrected_theory) / z4_corrected_theory
print(f"  Error: {error_z4_pct:.2f}%")
print()

# Test S⁶
print("Testing S⁶ (7D ambient space)...")
z6_meas, z6_err, d6_sq = test_s6_geodesics(trials=30000)
z6_theory = zeta_6()
z6_corrected_theory = np.sqrt(7) * z6_theory

print(f"  E[distance²] = {d6_sq:.6f}")
print(f"  ζ(6) measured = {z6_meas:.6f} ± {z6_err:.6f}")
print(f"  ζ(6) theory = {z6_theory:.6f}")
print(f"  √7 × ζ(6) = {z6_corrected_theory:.6f}")
error_z6_pct = 100 * abs(z6_meas - z6_corrected_theory) / z6_corrected_theory
print(f"  Error: {error_z6_pct:.2f}%")
print()

# ============================================================================
# VISUALIZATION
# ============================================================================

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot results
dimensions = [3, 5, 7]
measured = [z2_meas, z4_meas, z6_meas]
predicted = [z2_corrected_theory, z4_corrected_theory, z6_corrected_theory]
errors = [z2_err, z4_err, z6_err]

ax = axes[0]
x = np.arange(len(dimensions))
width = 0.35
ax.bar(x - width/2, measured, width, label='Measured', alpha=0.8)
ax.bar(x + width/2, predicted, width, label='Theory (√n × ζ)', alpha=0.8)
ax.set_ylabel('Zeta Value')
ax.set_xlabel('Ambient Dimension')
ax.set_xticks(x)
ax.set_xticklabels([f'R^{d}' for d in dimensions])
ax.legend()
ax.set_title('Measured vs Predicted Zeta Values')
ax.grid(alpha=0.3)

# Plot errors
ax = axes[1]
errors_pct = [error_z2_pct, error_z4_pct, error_z6_pct]
ax.bar(range(len(dimensions)), errors_pct, color=['green', 'orange', 'red'], alpha=0.7)
ax.set_ylabel('Error (%)')
ax.set_xlabel('Ambient Dimension')
ax.set_xticks(range(len(dimensions)))
ax.set_xticklabels([f'R^{d}' for d in dimensions])
ax.set_title('Prediction Error vs Dimension')
ax.axhline(y=5, color='red', linestyle='--', label='5% threshold', alpha=0.5)
ax.grid(alpha=0.3)

# Plot the dimensional pattern
ax = axes[2]
sqrt_dims = np.array([np.sqrt(3), np.sqrt(5), np.sqrt(7)])
ratios = np.array(measured) / np.array([zeta_2(), zeta_4(), zeta_6()])
ax.scatter(sqrt_dims, ratios, s=100, alpha=0.7)
ax.plot(sqrt_dims, sqrt_dims, 'r--', label='y = √n (hypothesis)', linewidth=2)
ax.set_xlabel('√(Ambient Dimension)')
ax.set_ylabel('Measured / Bare Zeta')
ax.set_title('Dimensional Scaling Pattern')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('zeta_dimensional_test.png', dpi=100, bbox_inches='tight')
plt.show()

print("=" * 80)
print("HYPOTHESIS CHECK:")
print("If ζ(2n) = √(2n+1) × [bare ζ(2n)], then measured/theory ratios should ≈ √n")
print(f"  S² ratio: {measured[0]/z2_theory:.4f} vs √3 = {np.sqrt(3):.4f}")
print(f"  S⁴ ratio: {measured[1]/z4_theory:.4f} vs √5 = {np.sqrt(5):.4f}")
print(f"  S⁶ ratio: {measured[2]/z6_theory:.4f} vs √7 = {np.sqrt(7):.4f}")
print("=" * 80)
````
