```python
# PASTE THIS IN A NEW COLAB CELL → CLICK "RUN"
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# CONFIG: YOUR UNIVERSE RULES (CHANGE THESE!)
grid_size = 20      # Pixel count (try 10 → 50)
phase_speed = 0.1   # How fast arrows spin (try 0.01 → 1.0)
collapse_radius = 3 # How far "measurement" spreads (try 1 → 10)

# STEP 1: CREATE YOUR EULER PIXEL UNIVERSE
np.random.seed(42)  # Makes results repeatable
phases = np.random.uniform(0, 2*np.pi, (grid_size, grid_size))  # Random arrow directions

# STEP 2: DEFINE "SNAPSHOTS" (QUANTUM MEASUREMENTS)
def take_snapshot():
    # Euler's magic: e^{iθ} = cosθ + i sinθ → we measure the REAL part (position)
    return np.cos(phases)

# STEP 3: SIMULATE "COLLAPSE" (CLICK TO MEASURE!)
def collapse(x, y):
    y, x = int(y), int(x)  # Fix axis order
    for i in range(max(0, x-collapse_radius), min(grid_size, x+collapse_radius)):
        for j in range(max(0, y-collapse_radius), min(grid_size, y+collapse_radius)):
            # FORCE PIXELS TO SYNC PHASES (like your crayon experiment!)
            phases[j, i] = phases[y, x]

# STEP 4: ANIMATE YOUR QUANTUM UNIVERSE
fig, ax = plt.subplots(figsize=(6,6))
im = ax.imshow(take_snapshot(), cmap='bwr', animated=True)
plt.title("Your Euler Pixel Universe (RED = +1, BLUE = -1)")

def update(frame):
    global phases
    phases += phase_speed  # SPIN ALL ARROWS!
    if frame % 15 == 0:   # Every 15 frames, "measure" at center
        collapse(grid_size//2, grid_size//2)
    im.set_array(take_snapshot())
    return [im]

ani = FuncAnimation(fig, update, frames=200, interval=50, blit=True)
plt.show()
```
