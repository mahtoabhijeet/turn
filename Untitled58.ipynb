```python
# ==============================================================================
# 1. SETUP: Install necessary libraries (if not in Colab)
# ==============================================================================
# !pip install torch numpy matplotlib

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display
import io
import base64

# Use GPU if available (vital for this)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ==============================================================================
# 2. THE "HARMONY" (The Target)
# This is the "stable attractor" or "harmony" we want to emerge.
# ==============================================================================
GRID_SIZE = 64
CHANNELS = 16  # 4 for RGBA, 12 for hidden "eigenmodes"

def get_target_smiley(size=GRID_SIZE):
    """Creates a 4-channel (RGBA) smiley face as the target 'harmony'."""
    # Create a blank canvas (H, W, C)
    canvas = np.zeros((size, size, 4), dtype=np.float32)

    # Face (yellow circle)
    center_x, center_y, radius = size // 2, size // 2, size // 3
    y, x = np.ogrid[:size, :size]
    mask = (x - center_x)**2 + (y - center_y)**2 < radius**2
    canvas[mask, 0] = 1.0  # R
    canvas[mask, 1] = 1.0  # G
    canvas[mask, 3] = 1.0  # Alpha

    # Left eye
    y, x = np.ogrid[:size, :size]
    mask = (x - (center_x - radius//2))**2 + (y - (center_y - radius//3))**2 < (radius//6)**2
    canvas[mask, 0:3] = 0.0  # Black

    # Right eye
    y, x = np.ogrid[:size, :size]
    mask = (x - (center_x + radius//2))**2 + (y - (center_y - radius//3))**2 < (radius//6)**2
    canvas[mask, 0:3] = 0.0  # Black

    # Smile
    y, x = np.ogrid[:size, :size]
    y_smile = center_y + radius//3
    x_smile = x - center_x
    mask1 = (x_smile**2 + (y - y_smile)**2 < (radius//1.5)**2)
    mask2 = (y > y_smile)
    canvas[mask1 & mask2, 0:3] = 0.0 # Black

    # Convert to PyTorch tensor: (Batch, Channels, Height, Width)
    target_tensor = torch.tensor(canvas).permute(2, 0, 1).unsqueeze(0)

    # Pad the target to match the full 16-channel state
    padding = torch.zeros(1, CHANNELS - 4, size, size)
    return torch.cat([target_tensor, padding], dim=1).to(device)

# ==============================================================================
# 3. THE "LATTICE" and "LOCAL COUPLING"
# The NCA model:
# - "Lattice" = The grid of cells
# - "Local Coupling" = A fixed perception filter (Sobel filters)
# - "Rule" = The *trainable* (twistable) part that interprets the local state
# ==============================================================================

class NCA(nn.Module):
    def __init__(self, in_channels=CHANNELS, hidden_channels=64, out_channels=CHANNELS):
        super().__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels

        # -----------------------------------------------------------------
        # The "Local Coupling" (Fixed Physics)
        # We define a *fixed* set of 3x3 convolution kernels (sensors)
        # to "perceive" the local neighborhood. This is not trained.
        # This is the "physics" of our toy universe.
        # -----------------------------------------------------------------

        # 1. Identity filter (what is my own state?)
        identity = torch.tensor([[0,0,0],[0,1,0],[0,0,0]], dtype=torch.float32)
        # 2. Sobel X filter (what's the gradient to my left/right?)
        sobel_x = torch.tensor([[-1,0,1],[-2,0,2],[-1,0,1]], dtype=torch.float32)
        # 3. Sobel Y filter (what's the gradient above/below me?)
        sobel_y = torch.tensor([[-1,-2,-1],[0,0,0],[1,2,1]], dtype=torch.float32)

        # Stack them: [3, 3, 3] -> [3, 1, 3, 3]
        all_kernels = torch.stack([identity, sobel_x, sobel_y]).unsqueeze(1)

        # Repeat these 3 filters for *every channel* independently
        # Final shape: [CHANNELS * 3, 1, 3, 3]
        all_kernels = all_kernels.repeat(self.in_channels, 1, 1, 1)

        # Create the fixed perception layer
        # `groups=in_channels` makes it a "depthwise" convolution,
        # so each channel is processed independently by the 3 filters.
        self.perception = nn.Conv2d(
            in_channels,
            in_channels * 3,
            kernel_size=3,
            padding=1,
            groups=in_channels,
            bias=False
        )
        self.perception.weight.data = all_kernels
        self.perception.weight.requires_grad = False  # This is KEY. We don't train the "physics".

        # -----------------------------------------------------------------
        # The "Rule" (The "Twistable" / Trainable Part)
        # This is a tiny neural net that interprets the sensor readings
        # from the perception layer and decides what the cell's next state should be.
        # This is the *only* part we will "twist" (train).
        # -----------------------------------------------------------------
        self.rule = nn.Sequential(
            nn.Conv2d(in_channels * 3, hidden_channels, kernel_size=1),
            nn.ReLU(),
            nn.Conv2d(hidden_channels, out_channels, kernel_size=1)
        )

        # Initialize the output layer to zeros.
        # This makes the "delta" (update) zero at the start,
        # which is a stable place to begin "twisting".
        self.rule[-1].weight.data.fill_(0.0)
        self.rule[-1].bias.data.fill_(0.0)

    def forward(self, x):
        # 1. "Local Coupling": Each cell "perceives" its neighborhood.
        perceived_state = self.perception(x)

        # 2. "The Rule": Apply the "twistable" logic to the perceived state
        # to calculate the "delta" (the update).
        delta = self.rule(perceived_state)

        # 3. Stochastic Update: Only update a random 50% of cells.
        # This makes the "harmony" more robust, as it must learn
        # to "repair" itself from this constant noise.
        update_mask = (torch.rand_like(x[:, 0:1, :, :]) < 0.5)

        # 4. "Living" Cells: Only "living" cells (alpha > 0.1) can update.
        # This is the "GF(2)" on/off state.
        living_mask = (F.max_pool2d(x[:, 0:1, :, :], kernel_size=3, stride=1, padding=1) > 0.1)

        # 5. Apply the update
        x = x + delta * update_mask * living_mask

        # 6. Clamp values: Ensure all "frequencies" stay between 0 and 1.
        x = x.clamp(0.0, 1.0)

        return x

# ==============================================================================
# 4. UTILITIES: Animation (To see the "results") - CORRECTED (AGAIN)
# ==============================================================================

def animate_rollout(rollout_frames, title=""):
    """Creates an HTML5 video animation of the growth process."""
    rgba_frames = [frame[:4].permute(1, 2, 0).cpu().detach().numpy() for frame in rollout_frames]

    fig, ax = plt.subplots()
    ax.set_title(title)
    ax.axis('off')
    img = ax.imshow(rgba_frames[0].clip(0, 1))

    def update(frame_idx):
        img.set_data(rgba_frames[frame_idx].clip(0, 1))
        return [img]

    ani = FuncAnimation(fig, update, frames=len(rgba_frames), interval=50, blit=True)
    plt.close(fig) # Prevent static image from displaying

    # --- THIS IS THE CORRECTED SECTION ---

    # Define a temporary file path in Colab's environment
    temp_filename = '/tmp/temp_animation.mp4'

    # 1. Save to the temporary file *on disk*
    ani.save(temp_filename, writer='ffmpeg', extra_args=['-vcodec', 'libx264', '-pix_fmt', 'yuv420p'])

    # 2. Read that file's contents back into an in-memory buffer
    buf = io.BytesIO()
    with open(temp_filename, 'rb') as f:
        buf.write(f.read())
    buf.seek(0)

    # --- END CORRECTED SECTION ---

    video_data = base64.b64encode(buf.read()).decode()

    return HTML(f'<video width=256 controls loop autoplay src="data:video/mp4;base6F4,{video_data}">')

# ==============================================================================
# 5. THE "TWISTING" (The Training Loop)
# ==============================================================================

print("Starting the 'twisting' (training) process...")

# --- Parameters ---
BATCH_SIZE = 8
PHYSICS_STEPS = 96   # How long the "music" is allowed to play (growth steps)
TRAINING_STEPS = 3001 # How many times we "twist" the rule
LEARNING_RATE = 1e-3

# --- Initialization ---
model = NCA().to(device)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# The "harmony" we're trying to achieve
target_harmony = get_target_smiley()
target_harmony = target_harmony.repeat(BATCH_SIZE, 1, 1, 1) # Create a batch of targets

# The "seed" (our "GF(2)" bit)
# A single "on" pixel in the center of the grid.
seed = torch.zeros(BATCH_SIZE, CHANNELS, GRID_SIZE, GRID_SIZE, device=device)
seed[:, :, GRID_SIZE // 2, GRID_SIZE // 2] = 1.0

# --- The Loop ---
for step in range(TRAINING_STEPS):

    # "Play the music": Grow the pattern from the seed for N physics steps
    state = seed.clone()
    rollout_for_viz = [] # Store frames for animation

    for i in range(PHYSICS_STEPS):
        state = model(state)
        if step % 200 == 0 and i % 4 == 0: # Save frames every 200 training steps
            rollout_for_viz.append(state[0].detach()) # Save 1st batch item

    # "Calculate Dissonance" (The Loss)
    # How far is our final "music" from the target "harmony"?
    # We only care about the visible RGBA channels.
    loss = F.mse_loss(state[:, :4], target_harmony[:, :4])

    # "The Twist" (Gradient Descent)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # "The Results" (Visualization)
    if step % 200 == 0:
        print(f"--- Step {step} --- Loss: {loss.item():.6f}")

        # Display the animation of the growth process
        anim = animate_rollout(rollout_for_viz, title=f"Growth at Training Step {step}")
        display(anim)

        # Display the final frame vs the target
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3))
        ax1.imshow(state[0, :4].permute(1, 2, 0).cpu().detach().numpy().clip(0, 1))
        ax1.set_title("Final State")
        ax1.axis('off')

        ax2.imshow(target_harmony[0, :4].permute(1, 2, 0).cpu().detach().numpy().clip(0, 1))
        ax2.set_title("Target Harmony")
        ax2.axis('off')

        plt.show()

print("Training complete. You can now analyze the emergent 'music'.")
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.linalg import null_space
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import itertools

class GF2Grid:
    """Graph representation of ARC grid with GF(2) Laplacian"""

    def __init__(self, grid, connectivity='4'):
        """
        grid: 2D numpy array of color values
        connectivity: '4' or '8' for 4-connected or 8-connected
        """
        self.grid = grid.astype(int)
        self.h, self.w = grid.shape
        self.n_nodes = self.h * self.w
        self.connectivity = connectivity
        self.A = self._build_adjacency()
        self.D = self._build_degree()
        self.L = self._build_laplacian()

    def _idx_2d_to_1d(self, i, j):
        """Convert 2D grid coordinates to 1D node index"""
        return i * self.w + j

    def _idx_1d_to_2d(self, idx):
        """Convert 1D node index to 2D grid coordinates"""
        return idx // self.w, idx % self.w

    def _build_adjacency(self):
        """Build adjacency matrix A over GF(2)"""
        A = np.zeros((self.n_nodes, self.n_nodes), dtype=int)

        for i in range(self.h):
            for j in range(self.w):
                u = self._idx_2d_to_1d(i, j)

                # 4-connectivity: up, down, left, right
                neighbors = [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]

                # 8-connectivity: add diagonals
                if self.connectivity == '8':
                    neighbors.extend([(i-1, j-1), (i-1, j+1), (i+1, j-1), (i+1, j+1)])

                for ni, nj in neighbors:
                    if 0 <= ni < self.h and 0 <= nj < self.w:
                        v = self._idx_2d_to_1d(ni, nj)
                        A[u, v] = 1

        return A

    def _build_degree(self):
        """Build degree matrix D over GF(2)"""
        degrees = self.A.sum(axis=1).astype(int) % 2  # mod 2
        return np.diag(degrees)

    def _build_laplacian(self):
        """Build Laplacian L = D XOR A over GF(2)"""
        L = (self.D + self.A) % 2  # XOR is addition mod 2
        return L

    def apply_automat_step(self, state, weights, bias):
        """
        Single XOR automaton step: s(t+1) = bias XOR (L . s(t) . weights)

        state: 1D array of current node states
        weights: 1D array of XOR weights (0 or 1) for Laplacian contribution
        bias: 1D array of bias terms
        """
        # L . s in GF(2)
        laplacian_effect = (self.L @ state) % 2

        # Apply weights (element-wise XOR)
        weighted = (laplacian_effect * weights) % 2

        # Add bias (XOR)
        next_state = (weighted + bias) % 2

        return next_state

    def diffuse_xor(self, state, steps=10):
        """
        Pure diffusion without learned weights: s(t+1) = bias XOR (L . s(t))
        Used for simple propagation tasks
        """
        s = state.copy()
        for _ in range(steps):
            s = (self.L @ s) % 2
        return s


class XORAutomat:
    """Learnable XOR automaton for ARC transformations"""

    def __init__(self, gf2_grid, n_steps=5):
        self.gf2 = gf2_grid
        self.n_steps = n_steps
        self.n_nodes = gf2_grid.n_nodes

        # Learn weight and bias for each node
        self.weights = np.random.randint(0, 2, self.n_nodes)
        self.bias = np.random.randint(0, 2, self.n_nodes)

    def apply(self, state):
        """Apply automat​on for n_steps iterations"""
        s = state.copy().astype(int)
        trajectory = [s.copy()]

        for _ in range(self.n_steps):
            s = self.gf2.apply_automat_step(s, self.weights, self.bias)
            trajectory.append(s.copy())

        return s, trajectory

    def loss(self, input_state, target_state):
        """Compute XOR loss: count mismatched bits"""
        output, _ = self.apply(input_state)
        return np.sum(np.abs(output - target_state))

    def fit_constraint_solver(self, input_state, target_state):
        """
        Brute force search for weights/bias over small neighborhoods
        Practical for simple tasks; use SGD for larger problems
        """
        best_loss = float('inf')
        best_w, best_b = self.weights.copy(), self.bias.copy()

        # Sample random configurations
        for _ in range(1000):
            test_w = np.random.randint(0, 2, self.n_nodes)
            test_b = np.random.randint(0, 2, self.n_nodes)

            self.weights, self.bias = test_w, test_b
            loss = self.loss(input_state, target_state)

            if loss < best_loss:
                best_loss = loss
                best_w, best_b = test_w.copy(), test_b.copy()

        self.weights, self.bias = best_w, best_b
        return best_loss

    def fit_sgd(self, input_state, target_state, lr=0.01, iters=100):
        """
        Gradient-free SGD using finite differences over GF(2)
        Flip bits and measure loss change
        """
        losses = []

        for iteration in range(iters):
            current_loss = self.loss(input_state, target_state)
            losses.append(current_loss)

            # Try flipping each weight and bias bit
            for idx in range(self.n_nodes):
                # Try flipping weight
                self.weights[idx] ^= 1
                new_loss_w = self.loss(input_state, target_state)
                if new_loss_w > current_loss:
                    self.weights[idx] ^= 1  # flip back

                # Try flipping bias
                self.bias[idx] ^= 1
                new_loss_b = self.loss(input_state, target_state)
                if new_loss_b > current_loss:
                    self.bias[idx] ^= 1  # flip back

        return losses


# ============= SIMPLE ARC TASK GENERATORS =============

def task_fill_enclosed_region():
    """Task: Fill interior of a rectangular boundary with a color"""
    # Create boundary (color 1) around edges
    grid_in = np.zeros((8, 8), dtype=int)
    grid_in[1:7, 1:7] = 0  # Empty interior
    grid_in[0, :] = 1
    grid_in[7, :] = 1
    grid_in[:, 0] = 1
    grid_in[:, 7] = 1

    # Expected output: filled interior
    grid_out = grid_in.copy()
    grid_out[2:6, 2:6] = 2  # Fill with color 2

    return grid_in, grid_out


def task_propagate_color():
    """Task: Propagate color from source through connected component"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[2, 2] = 2  # Source

    grid_out = np.zeros((6, 6), dtype=int)
    grid_out[1:4, 1:4] = 2  # Propagate to 3x3 region

    return grid_in, grid_out


def task_symmetry_completion():
    """Task: Complete symmetric pattern"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[2, 2] = 1  # Center
    grid_in[1, 2] = 1  # Top
    grid_in[2, 1] = 1  # Left

    grid_out = np.zeros((5, 5), dtype=int)
    grid_out[2, 2] = 1  # Center
    grid_out[1, 2] = 1  # Top
    grid_out[2, 1] = 1  # Left
    grid_out[3, 2] = 1  # Bottom (mirror)
    grid_out[2, 3] = 1  # Right (mirror)

    return grid_in, grid_out


# ============= VISUALIZATION =============

def visualize_grid(grid, title=""):
    """Display grid with colors"""
    cmap = ListedColormap(['white', 'blue', 'red', 'green', 'yellow', 'cyan'])
    plt.figure(figsize=(6, 6))
    plt.imshow(grid, cmap=cmap, vmin=0, vmax=5)
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.xticks([])
    plt.yticks([])


def compare_grids(grid_in, grid_truth, grid_pred, task_name=""):
    """Compare input, ground truth, and prediction"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    cmap = ListedColormap(['white', 'blue', 'red', 'green', 'yellow', 'cyan'])

    axes[0].imshow(grid_in, cmap=cmap, vmin=0, vmax=5)
    axes[0].set_title("Input")
    axes[0].grid(True, alpha=0.3)

    axes[1].imshow(grid_truth, cmap=cmap, vmin=0, vmax=5)
    axes[1].set_title("Ground Truth")
    axes[1].grid(True, alpha=0.3)

    axes[2].imshow(grid_pred, cmap=cmap, vmin=0, vmax=5)
    axes[2].set_title("Prediction")
    axes[2].grid(True, alpha=0.3)

    accuracy = np.mean(grid_truth == grid_pred)
    fig.suptitle(f"{task_name} | Accuracy: {accuracy:.2%}", fontsize=14)
    plt.tight_layout()


# ============= MAIN EXPERIMENTS =============

def run_experiment(grid_in, grid_out, task_name, use_diffusion=False, use_sgd=False):
    """Run ARC task with XOR automaton"""
    print(f"\n{'='*60}")
    print(f"Task: {task_name}")
    print(f"{'='*60}")

    # Create GF(2) grid
    gf2 = GF2Grid(grid_in, connectivity='4')
    print(f"Grid size: {gf2.h}x{gf2.w} = {gf2.n_nodes} nodes")
    print(f"Laplacian rank: {np.linalg.matrix_rank(gf2.L)}")

    # Flatten for processing
    input_state = grid_in.flatten()
    target_state = grid_out.flatten()

    if use_diffusion:
        # Pure diffusion without learning
        print("\nUsing pure XOR diffusion...")
        output_state = gf2.diffuse_xor(input_state, steps=15)
    else:
        # Learn XOR automaton
        automat = XORAutomat(gf2, n_steps=5)

        if use_sgd:
            print("\nLearning with SGD...")
            losses = automat.fit_sgd(input_state, target_state, iters=50)
            print(f"Initial loss: {losses[0]}, Final loss: {losses[-1]}")
        else:
            print("\nLearning with constraint solver...")
            loss = automat.fit_constraint_solver(input_state, target_state)
            print(f"Best loss found: {loss}")

        output_state, trajectory = automat.apply(input_state)

    # Reshape and evaluate
    output_grid = output_state.reshape(grid_in.shape)
    accuracy = np.mean(grid_out == output_grid)

    print(f"Accuracy: {accuracy:.2%}")
    print(f"Mismatched cells: {np.sum(grid_out != output_grid)}")

    # Visualize
    compare_grids(grid_in, grid_out, output_grid, task_name)
    plt.show()

    return output_grid


if __name__ == "__main__":
    # Task 1: Fill enclosed region
    grid_in, grid_out = task_fill_enclosed_region()
    run_experiment(grid_in, grid_out, "Fill Enclosed Region", use_diffusion=True)

    # Task 2: Propagate color
    grid_in, grid_out = task_propagate_color()
    run_experiment(grid_in, grid_out, "Propagate Color", use_diffusion=True)

    # Task 3: Symmetry completion (requires learning)
    grid_in, grid_out = task_symmetry_completion()
    run_experiment(grid_in, grid_out, "Symmetry Completion", use_sgd=True)
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.linalg import null_space
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import itertools

class GF2Grid:
    """Graph representation of ARC grid with GF(2) Laplacian"""

    def __init__(self, grid, connectivity='4'):
        """
        grid: 2D numpy array of color values
        connectivity: '4' or '8' for 4-connected or 8-connected
        """
        self.grid = grid.astype(int)
        self.h, self.w = grid.shape
        self.n_nodes = self.h * self.w
        self.connectivity = connectivity
        self.A = self._build_adjacency()
        self.D = self._build_degree()
        self.L = self._build_laplacian()

    def _idx_2d_to_1d(self, i, j):
        """Convert 2D grid coordinates to 1D node index"""
        return i * self.w + j

    def _idx_1d_to_2d(self, idx):
        """Convert 1D node index to 2D grid coordinates"""
        return idx // self.w, idx % self.w

    def _build_adjacency(self):
        """Build adjacency matrix A over GF(2)"""
        A = np.zeros((self.n_nodes, self.n_nodes), dtype=int)

        for i in range(self.h):
            for j in range(self.w):
                u = self._idx_2d_to_1d(i, j)

                neighbors = [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]

                if self.connectivity == '8':
                    neighbors.extend([(i-1, j-1), (i-1, j+1), (i+1, j-1), (i+1, j+1)])

                for ni, nj in neighbors:
                    if 0 <= ni < self.h and 0 <= nj < self.w:
                        v = self._idx_2d_to_1d(ni, nj)
                        A[u, v] = 1

        return A

    def _build_degree(self):
        """Build degree matrix D over GF(2)"""
        degrees = self.A.sum(axis=1).astype(int) % 2
        return np.diag(degrees)

    def _build_laplacian(self):
        """Build Laplacian L = D XOR A over GF(2)"""
        L = (self.D + self.A) % 2
        return L

    def apply_automat_step(self, state, weights, bias):
        """Single XOR automaton step over GF(2)"""
        laplacian_effect = (self.L @ state) % 2
        weighted = (laplacian_effect * weights) % 2
        next_state = (weighted + bias) % 2
        return next_state

    def diffuse_xor(self, state, steps=10):
        """Pure diffusion: s(t+1) = L . s(t) mod 2"""
        s = state.copy()
        for _ in range(steps):
            s = (self.L @ s) % 2
        return s


class MultiChannelXORAutomat:
    """XOR automaton with separate channels per color"""

    def __init__(self, gf2_grid, n_colors=3, n_steps=5):
        self.gf2 = gf2_grid
        self.n_colors = n_colors
        self.n_steps = n_steps
        self.n_nodes = gf2_grid.n_nodes

        # Per-color learnable weights and bias
        self.weights = np.random.randint(0, 2, (n_colors, self.n_nodes))
        self.bias = np.random.randint(0, 2, (n_colors, self.n_nodes))

    def apply(self, state_channels):
        """
        state_channels: (n_colors, n_nodes) array
        Each channel is processed independently through XOR automat
        Output is the argmax color at each node
        """
        channels = state_channels.copy().astype(int)

        for step in range(self.n_steps):
            new_channels = np.zeros_like(channels)
            for c in range(self.n_colors):
                new_channels[c] = self.gf2.apply_automat_step(
                    channels[c], self.weights[c], self.bias[c]
                )
            channels = new_channels

        # Decode: argmax across channels
        output = np.argmax(channels, axis=0)
        return output, channels

    def loss(self, input_channels, target_grid):
        """Loss = mismatched cells after decoding"""
        output, _ = self.apply(input_channels)
        return np.sum(output != target_grid)

    def fit_sgd(self, input_channels, target_grid, iters=100):
        """Gradient-free SGD over channel weights and biases"""
        losses = []

        for iteration in range(iters):
            current_loss = self.loss(input_channels, target_grid)
            losses.append(current_loss)

            # Try flipping each weight and bias
            for c in range(self.n_colors):
                for idx in range(self.n_nodes):
                    # Flip weight
                    self.weights[c, idx] ^= 1
                    new_loss_w = self.loss(input_channels, target_grid)
                    if new_loss_w > current_loss:
                        self.weights[c, idx] ^= 1
                    else:
                        current_loss = new_loss_w

                    # Flip bias
                    self.bias[c, idx] ^= 1
                    new_loss_b = self.loss(input_channels, target_grid)
                    if new_loss_b > current_loss:
                        self.bias[c, idx] ^= 1
                    else:
                        current_loss = new_loss_b

        return losses


def grid_to_channels(grid, n_colors):
    """Convert grid to one-hot encoded channels"""
    h, w = grid.shape
    channels = np.zeros((n_colors, h * w), dtype=int)
    flat_grid = grid.flatten()
    for c in range(n_colors):
        channels[c] = (flat_grid == c).astype(int)
    return channels


def channels_to_grid(channels):
    """Convert channels back to grid via argmax"""
    return np.argmax(channels, axis=0)


# ============= IMPROVED TASK GENERATORS =============

def task_fill_enclosed_region():
    """Fill interior of rectangular boundary"""
    grid_in = np.zeros((8, 8), dtype=int)
    grid_in[0, :] = 1
    grid_in[7, :] = 1
    grid_in[:, 0] = 1
    grid_in[:, 7] = 1

    grid_out = grid_in.copy()
    grid_out[2:6, 2:6] = 2

    return grid_in, grid_out


def task_propagate_color():
    """Propagate color from source through 4-connected component"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[2, 2] = 2

    grid_out = np.zeros((6, 6), dtype=int)
    grid_out[1:4, 1:4] = 2

    return grid_in, grid_out


def task_symmetry_completion():
    """Complete symmetric pattern"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[2, 2] = 1
    grid_in[1, 2] = 1
    grid_in[2, 1] = 1

    grid_out = np.zeros((5, 5), dtype=int)
    grid_out[2, 2] = 1
    grid_out[1, 2] = 1
    grid_out[2, 1] = 1
    grid_out[3, 2] = 1
    grid_out[2, 3] = 1

    return grid_in, grid_out


def task_connected_component_fill():
    """Fill all connected non-zero cells with target color"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[2:4, 2:4] = 1  # 2x2 block
    grid_in[2, 4] = 1      # Connected extension

    grid_out = np.ones((6, 6), dtype=int) * 2
    grid_out[2:4, 2:4] = 2
    grid_out[2, 4] = 2

    return grid_in, grid_out


def task_invert_nonzero():
    """Invert all nonzero cells"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[1:4, 1:4] = 1
    grid_in[2, 2] = 0

    grid_out = np.zeros((5, 5), dtype=int)
    grid_out[1:4, 1:4] = 2
    grid_out[2, 2] = 0

    return grid_in, grid_out


# ============= VISUALIZATION =============

def compare_grids(grid_in, grid_truth, grid_pred, task_name=""):
    """Compare input, ground truth, and prediction"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    cmap = ListedColormap(['white', 'blue', 'red', 'green', 'yellow', 'cyan', 'magenta'])

    axes[0].imshow(grid_in, cmap=cmap, vmin=0, vmax=6)
    axes[0].set_title("Input")
    axes[0].grid(True, alpha=0.3)

    axes[1].imshow(grid_truth, cmap=cmap, vmin=0, vmax=6)
    axes[1].set_title("Ground Truth")
    axes[1].grid(True, alpha=0.3)

    axes[2].imshow(grid_pred, cmap=cmap, vmin=0, vmax=6)
    axes[2].set_title("Prediction")
    axes[2].grid(True, alpha=0.3)

    accuracy = np.mean(grid_truth == grid_pred)
    fig.suptitle(f"{task_name} | Accuracy: {accuracy:.2%}", fontsize=14)
    plt.tight_layout()


# ============= MAIN EXPERIMENTS =============

def run_experiment(grid_in, grid_out, task_name, use_learning=True, n_colors=3):
    """Run ARC task with multi-channel XOR automaton"""
    print(f"\n{'='*60}")
    print(f"Task: {task_name}")
    print(f"{'='*60}")

    gf2 = GF2Grid(grid_in, connectivity='4')
    print(f"Grid size: {gf2.h}x{gf2.w} = {gf2.n_nodes} nodes")
    print(f"Laplacian rank: {np.linalg.matrix_rank(gf2.L)}")
    print(f"Colors in output: {np.unique(grid_out)}")

    # Convert to multi-channel representation
    input_channels = grid_to_channels(grid_in, n_colors)

    if use_learning:
        print("\nLearning with multi-channel SGD...")
        automat = MultiChannelXORAutomat(gf2, n_colors=n_colors, n_steps=5)
        losses = automat.fit_sgd(input_channels, grid_out, iters=80)
        print(f"Initial loss: {losses[0]}, Final loss: {losses[-1]}")

        output_grid, _ = automat.apply(input_channels)
    else:
        print("\nUsing learned initialization...")
        output_grid = grid_in.copy()

    accuracy = np.mean(grid_out == output_grid)
    print(f"Accuracy: {accuracy:.2%}")
    print(f"Mismatched cells: {np.sum(grid_out != output_grid)}")

    compare_grids(grid_in, grid_out, output_grid, task_name)
    plt.show()

    return output_grid


if __name__ == "__main__":
    # Task 1: Fill enclosed region (needs learning)
    grid_in, grid_out = task_fill_enclosed_region()
    run_experiment(grid_in, grid_out, "Fill Enclosed Region", use_learning=True, n_colors=3)

    # Task 2: Propagate color (needs learning)
    grid_in, grid_out = task_propagate_color()
    run_experiment(grid_in, grid_out, "Propagate Color", use_learning=True, n_colors=3)

    # Task 3: Symmetry completion
    grid_in, grid_out = task_symmetry_completion()
    run_experiment(grid_in, grid_out, "Symmetry Completion", use_learning=True, n_colors=2)

    # Task 4: Connected component fill
    grid_in, grid_out = task_connected_component_fill()
    run_experiment(grid_in, grid_out, "Connected Component Fill", use_learning=True, n_colors=3)

    # Task 5: Invert nonzero
    grid_in, grid_out = task_invert_nonzero()
    run_experiment(grid_in, grid_out, "Invert Nonzero", use_learning=True, n_colors=3)
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.linalg import null_space
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import itertools

class GF2Grid:
    """Graph representation of ARC grid with GF(2) Laplacian"""

    def __init__(self, grid, connectivity='4'):
        """
        grid: 2D numpy array of color values
        connectivity: '4' or '8' for 4-connected or 8-connected
        """
        self.grid = grid.astype(int)
        self.h, self.w = grid.shape
        self.n_nodes = self.h * self.w
        self.connectivity = connectivity
        self.A = self._build_adjacency()
        self.D = self._build_degree()
        self.L = self._build_laplacian()

    def _idx_2d_to_1d(self, i, j):
        """Convert 2D grid coordinates to 1D node index"""
        return i * self.w + j

    def _idx_1d_to_2d(self, idx):
        """Convert 1D node index to 2D grid coordinates"""
        return idx // self.w, idx % self.w

    def _build_adjacency(self):
        """Build adjacency matrix A over GF(2)"""
        A = np.zeros((self.n_nodes, self.n_nodes), dtype=int)

        for i in range(self.h):
            for j in range(self.w):
                u = self._idx_2d_to_1d(i, j)

                neighbors = [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]

                if self.connectivity == '8':
                    neighbors.extend([(i-1, j-1), (i-1, j+1), (i+1, j-1), (i+1, j+1)])

                for ni, nj in neighbors:
                    if 0 <= ni < self.h and 0 <= nj < self.w:
                        v = self._idx_2d_to_1d(ni, nj)
                        A[u, v] = 1

        return A

    def _build_degree(self):
        """Build degree matrix D over GF(2)"""
        degrees = self.A.sum(axis=1).astype(int) % 2
        return np.diag(degrees)

    def _build_laplacian(self):
        """Build Laplacian L = D XOR A over GF(2)"""
        L = (self.D + self.A) % 2
        return L

    def apply_automat_step(self, state, weights, bias):
        """Single XOR automaton step over GF(2)"""
        laplacian_effect = (self.L @ state) % 2
        weighted = (laplacian_effect * weights) % 2
        next_state = (weighted + bias) % 2
        return next_state

    def diffuse_xor(self, state, steps=10):
        """Pure diffusion: s(t+1) = L . s(t) mod 2"""
        s = state.copy()
        for _ in range(steps):
            s = (self.L @ s) % 2
        return s


class MultiChannelXORAutomat:
    """XOR automaton with separate channels per color"""

    def __init__(self, gf2_grid, n_colors=3, n_steps=5):
        self.gf2 = gf2_grid
        self.n_colors = n_colors
        self.n_steps = n_steps
        self.n_nodes = gf2_grid.n_nodes

        # Per-color learnable weights and bias
        self.weights = np.random.randint(0, 2, (n_colors, self.n_nodes))
        self.bias = np.random.randint(0, 2, (n_colors, self.n_nodes))

    def apply(self, state_channels):
        """
        state_channels: (n_colors, n_nodes) array
        Each channel is processed independently through XOR automat
        Output is the argmax color at each node
        """
        channels = state_channels.copy().astype(int)

        for step in range(self.n_steps):
            new_channels = np.zeros_like(channels)
            for c in range(self.n_colors):
                new_channels[c] = self.gf2.apply_automat_step(
                    channels[c], self.weights[c], self.bias[c]
                )
            channels = new_channels

        # Decode: argmax across channels
        output = np.argmax(channels, axis=0)
        return output, channels

    def loss(self, input_channels, target_grid):
        """Loss = mismatched cells after decoding"""
        output, _ = self.apply(input_channels)
        target_flat = target_grid.flatten()
        return np.sum(output != target_flat)

    def fit_sgd(self, input_channels, target_grid, iters=100):
        """Gradient-free SGD over channel weights and biases"""
        losses = []

        for iteration in range(iters):
            current_loss = self.loss(input_channels, target_grid)
            losses.append(current_loss)

            # Random sampling of improvements (faster than exhaustive search)
            n_candidates = min(100, self.n_nodes * self.n_colors // 4)

            for _ in range(n_candidates):
                c = np.random.randint(0, self.n_colors)
                idx = np.random.randint(0, self.n_nodes)

                # Try flipping weight
                self.weights[c, idx] ^= 1
                new_loss_w = self.loss(input_channels, target_grid)
                if new_loss_w > current_loss:
                    self.weights[c, idx] ^= 1
                else:
                    current_loss = new_loss_w

                # Try flipping bias
                self.bias[c, idx] ^= 1
                new_loss_b = self.loss(input_channels, target_grid)
                if new_loss_b > current_loss:
                    self.bias[c, idx] ^= 1
                else:
                    current_loss = new_loss_b

        return losses


def grid_to_channels(grid, n_colors):
    """Convert grid to one-hot encoded channels"""
    h, w = grid.shape
    channels = np.zeros((n_colors, h * w), dtype=int)
    flat_grid = grid.flatten()
    for c in range(n_colors):
        channels[c] = (flat_grid == c).astype(int)
    return channels


def channels_to_grid(channels):
    """Convert channels back to grid via argmax"""
    return np.argmax(channels, axis=0)


# ============= IMPROVED TASK GENERATORS =============

def task_fill_enclosed_region():
    """Fill interior of rectangular boundary"""
    grid_in = np.zeros((8, 8), dtype=int)
    grid_in[0, :] = 1
    grid_in[7, :] = 1
    grid_in[:, 0] = 1
    grid_in[:, 7] = 1

    grid_out = grid_in.copy()
    grid_out[2:6, 2:6] = 2

    return grid_in, grid_out


def task_propagate_color():
    """Propagate color from source through 4-connected component"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[2, 2] = 2

    grid_out = np.zeros((6, 6), dtype=int)
    grid_out[1:4, 1:4] = 2

    return grid_in, grid_out


def task_symmetry_completion():
    """Complete symmetric pattern"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[2, 2] = 1
    grid_in[1, 2] = 1
    grid_in[2, 1] = 1

    grid_out = np.zeros((5, 5), dtype=int)
    grid_out[2, 2] = 1
    grid_out[1, 2] = 1
    grid_out[2, 1] = 1
    grid_out[3, 2] = 1
    grid_out[2, 3] = 1

    return grid_in, grid_out


def task_connected_component_fill():
    """Fill all connected non-zero cells with target color"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[2:4, 2:4] = 1  # 2x2 block
    grid_in[2, 4] = 1      # Connected extension

    grid_out = np.ones((6, 6), dtype=int) * 2
    grid_out[2:4, 2:4] = 2
    grid_out[2, 4] = 2

    return grid_in, grid_out


def task_invert_nonzero():
    """Invert all nonzero cells"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[1:4, 1:4] = 1
    grid_in[2, 2] = 0

    grid_out = np.zeros((5, 5), dtype=int)
    grid_out[1:4, 1:4] = 2
    grid_out[2, 2] = 0

    return grid_in, grid_out


# ============= VISUALIZATION =============

def compare_grids(grid_in, grid_truth, grid_pred, task_name=""):
    """Compare input, ground truth, and prediction"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    cmap = ListedColormap(['white', 'blue', 'red', 'green', 'yellow', 'cyan', 'magenta'])

    axes[0].imshow(grid_in, cmap=cmap, vmin=0, vmax=6)
    axes[0].set_title("Input")
    axes[0].grid(True, alpha=0.3)

    axes[1].imshow(grid_truth, cmap=cmap, vmin=0, vmax=6)
    axes[1].set_title("Ground Truth")
    axes[1].grid(True, alpha=0.3)

    axes[2].imshow(grid_pred, cmap=cmap, vmin=0, vmax=6)
    axes[2].set_title("Prediction")
    axes[2].grid(True, alpha=0.3)

    accuracy = np.mean(grid_truth == grid_pred)
    fig.suptitle(f"{task_name} | Accuracy: {accuracy:.2%}", fontsize=14)
    plt.tight_layout()


# ============= MAIN EXPERIMENTS =============

def run_experiment(grid_in, grid_out, task_name, use_learning=True, n_colors=3):
    """Run ARC task with multi-channel XOR automaton"""
    print(f"\n{'='*60}")
    print(f"Task: {task_name}")
    print(f"{'='*60}")

    gf2 = GF2Grid(grid_in, connectivity='4')
    print(f"Grid size: {gf2.h}x{gf2.w} = {gf2.n_nodes} nodes")
    print(f"Laplacian rank: {np.linalg.matrix_rank(gf2.L)}")
    print(f"Colors in output: {np.unique(grid_out)}")

    # Convert to multi-channel representation
    input_channels = grid_to_channels(grid_in, n_colors)

    if use_learning:
        print("\nLearning with multi-channel SGD...")
        automat = MultiChannelXORAutomat(gf2, n_colors=n_colors, n_steps=5)
        losses = automat.fit_sgd(input_channels, grid_out, iters=80)
        print(f"Initial loss: {losses[0]}, Final loss: {losses[-1]}")

        output_grid, _ = automat.apply(input_channels)
        output_grid = output_grid.reshape(grid_in.shape)
    else:
        print("\nUsing learned initialization...")
        output_grid = grid_in.copy()

    accuracy = np.mean(grid_out == output_grid)
    print(f"Accuracy: {accuracy:.2%}")
    print(f"Mismatched cells: {np.sum(grid_out != output_grid)}")

    compare_grids(grid_in, grid_out, output_grid, task_name)
    plt.show()

    return output_grid


if __name__ == "__main__":
    # Task 1: Fill enclosed region (needs learning + more steps)
    grid_in, grid_out = task_fill_enclosed_region()
    gf2_1 = GF2Grid(grid_in)
    automat1 = MultiChannelXORAutomat(gf2_1, n_colors=3, n_steps=10)
    input_ch1 = grid_to_channels(grid_in, 3)
    losses1 = automat1.fit_sgd(input_ch1, grid_out, iters=150)
    output1, _ = automat1.apply(input_ch1)
    output1 = output1.reshape(grid_in.shape)
    compare_grids(grid_in, grid_out, output1, "Fill Enclosed Region")
    plt.show()

    # Task 2: Propagate color (more steps for longer range)
    grid_in, grid_out = task_propagate_color()
    gf2_2 = GF2Grid(grid_in)
    automat2 = MultiChannelXORAutomat(gf2_2, n_colors=3, n_steps=8)
    input_ch2 = grid_to_channels(grid_in, 3)
    losses2 = automat2.fit_sgd(input_ch2, grid_out, iters=150)
    output2, _ = automat2.apply(input_ch2)
    output2 = output2.reshape(grid_in.shape)
    compare_grids(grid_in, grid_out, output2, "Propagate Color")
    plt.show()

    # Task 3: Symmetry completion
    grid_in, grid_out = task_symmetry_completion()
    gf2_3 = GF2Grid(grid_in)
    automat3 = MultiChannelXORAutomat(gf2_3, n_colors=2, n_steps=5)
    input_ch3 = grid_to_channels(grid_in, 2)
    losses3 = automat3.fit_sgd(input_ch3, grid_out, iters=120)
    output3, _ = automat3.apply(input_ch3)
    output3 = output3.reshape(grid_in.shape)
    compare_grids(grid_in, grid_out, output3, "Symmetry Completion")
    plt.show()

    # Task 4: Connected component fill (more steps)
    grid_in, grid_out = task_connected_component_fill()
    gf2_4 = GF2Grid(grid_in)
    automat4 = MultiChannelXORAutomat(gf2_4, n_colors=3, n_steps=12)
    input_ch4 = grid_to_channels(grid_in, 3)
    losses4 = automat4.fit_sgd(input_ch4, grid_out, iters=150)
    output4, _ = automat4.apply(input_ch4)
    output4 = output4.reshape(grid_in.shape)
    compare_grids(grid_in, grid_out, output4, "Connected Component Fill")
    plt.show()

    # Task 5: Invert nonzero
    grid_in, grid_out = task_invert_nonzero()
    gf2_5 = GF2Grid(grid_in)
    automat5 = MultiChannelXORAutomat(gf2_5, n_colors=3, n_steps=5)
    input_ch5 = grid_to_channels(grid_in, 3)
    losses5 = automat5.fit_sgd(input_ch5, grid_out, iters=150)
    output5, _ = automat5.apply(input_ch5)
    output5 = output5.reshape(grid_in.shape)
    compare_grids(grid_in, grid_out, output5, "Invert Nonzero")
    plt.show()
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.linalg import null_space
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import itertools

class GF2Grid:
    """Graph representation of ARC grid with GF(2) Laplacian"""

    def __init__(self, grid, connectivity='4'):
        """
        grid: 2D numpy array of color values
        connectivity: '4' or '8' for 4-connected or 8-connected
        """
        self.grid = grid.astype(int)
        self.h, self.w = grid.shape
        self.n_nodes = self.h * self.w
        self.connectivity = connectivity
        self.A = self._build_adjacency()
        self.D = self._build_degree()
        self.L = self._build_laplacian()

    def _idx_2d_to_1d(self, i, j):
        return i * self.w + j

    def _idx_1d_to_2d(self, idx):
        return idx // self.w, idx % self.w

    def _build_adjacency(self):
        A = np.zeros((self.n_nodes, self.n_nodes), dtype=int)

        for i in range(self.h):
            for j in range(self.w):
                u = self._idx_2d_to_1d(i, j)
                neighbors = [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]

                if self.connectivity == '8':
                    neighbors.extend([(i-1, j-1), (i-1, j+1), (i+1, j-1), (i+1, j+1)])

                for ni, nj in neighbors:
                    if 0 <= ni < self.h and 0 <= nj < self.w:
                        v = self._idx_2d_to_1d(ni, nj)
                        A[u, v] = 1

        return A

    def _build_degree(self):
        degrees = self.A.sum(axis=1).astype(int) % 2
        return np.diag(degrees)

    def _build_laplacian(self):
        L = (self.D + self.A) % 2
        return L


class MultiChannelXORAutomat:
    """XOR automaton with separate channels per color"""

    def __init__(self, gf2_grid, n_colors=3, n_steps=5):
        self.gf2 = gf2_grid
        self.n_colors = n_colors
        self.n_steps = n_steps
        self.n_nodes = gf2_grid.n_nodes

        self.weights = np.random.randint(0, 2, (n_colors, self.n_nodes))
        self.bias = np.random.randint(0, 2, (n_colors, self.n_nodes))

    def apply(self, state_channels):
        channels = state_channels.copy().astype(int)

        for step in range(self.n_steps):
            new_channels = np.zeros_like(channels)
            for c in range(self.n_colors):
                laplacian_effect = (self.gf2.L @ channels[c]) % 2
                weighted = (laplacian_effect * self.weights[c]) % 2
                new_channels[c] = (weighted + self.bias[c]) % 2
            channels = new_channels

        output = np.argmax(channels, axis=0)
        return output, channels

    def loss(self, input_channels, target_grid):
        output, _ = self.apply(input_channels)
        target_flat = target_grid.flatten()
        return np.sum(output != target_flat)

    def fit_sgd(self, input_channels, target_grid, iters=100):
        losses = []

        for iteration in range(iters):
            current_loss = self.loss(input_channels, target_grid)
            losses.append(current_loss)

            n_candidates = min(100, self.n_nodes * self.n_colors // 4)

            for _ in range(n_candidates):
                c = np.random.randint(0, self.n_colors)
                idx = np.random.randint(0, self.n_nodes)

                self.weights[c, idx] ^= 1
                new_loss_w = self.loss(input_channels, target_grid)
                if new_loss_w > current_loss:
                    self.weights[c, idx] ^= 1
                else:
                    current_loss = new_loss_w

                self.bias[c, idx] ^= 1
                new_loss_b = self.loss(input_channels, target_grid)
                if new_loss_b > current_loss:
                    self.bias[c, idx] ^= 1
                else:
                    current_loss = new_loss_b

        return losses


def grid_to_channels(grid, n_colors):
    h, w = grid.shape
    channels = np.zeros((n_colors, h * w), dtype=int)
    flat_grid = grid.flatten()
    for c in range(n_colors):
        channels[c] = (flat_grid == c).astype(int)
    return channels


def compare_grids(grid_in, grid_truth, grid_pred, task_name=""):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    cmap = ListedColormap(['white', 'blue', 'red', 'green', 'yellow', 'cyan', 'magenta', 'gray'])

    axes[0].imshow(grid_in, cmap=cmap, vmin=0, vmax=7)
    axes[0].set_title("Input")
    axes[0].grid(True, alpha=0.3)

    axes[1].imshow(grid_truth, cmap=cmap, vmin=0, vmax=7)
    axes[1].set_title("Ground Truth")
    axes[1].grid(True, alpha=0.3)

    axes[2].imshow(grid_pred, cmap=cmap, vmin=0, vmax=7)
    axes[2].set_title("Prediction")
    axes[2].grid(True, alpha=0.3)

    accuracy = np.mean(grid_truth == grid_pred)
    fig.suptitle(f"{task_name} | Accuracy: {accuracy:.2%}", fontsize=14)
    plt.tight_layout()


def run_task(grid_in, grid_out, task_name, n_colors=3, n_steps=5, iters=150):
    print(f"\n{'='*70}")
    print(f"Task: {task_name}")
    print(f"{'='*70}")

    gf2 = GF2Grid(grid_in, connectivity='8')
    print(f"Grid: {gf2.h}x{gf2.w}, Colors in output: {np.unique(grid_out)}")

    input_ch = grid_to_channels(grid_in, n_colors)
    automat = MultiChannelXORAutomat(gf2, n_colors=n_colors, n_steps=n_steps)
    losses = automat.fit_sgd(input_ch, grid_out, iters=iters)
    output, _ = automat.apply(input_ch)
    output = output.reshape(grid_in.shape)

    accuracy = np.mean(grid_out == output)
    print(f"Accuracy: {accuracy:.2%} | Loss: {losses[-1]}")

    compare_grids(grid_in, grid_out, output, task_name)
    plt.show()

    return accuracy


# ============= DIVERSE TASK GENERATORS =============

def task_checkerboard():
    """Checkerboard pattern - XOR loves this"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_out = np.zeros((6, 6), dtype=int)
    for i in range(6):
        for j in range(6):
            grid_out[i, j] = (i + j) % 2
    return grid_in, grid_out


def task_diagonal_stripe():
    """Detect and mark diagonals"""
    grid_in = np.zeros((7, 7), dtype=int)
    for i in range(7):
        grid_in[i, i] = 1  # Main diagonal

    grid_out = np.zeros((7, 7), dtype=int)
    for i in range(7):
        grid_out[i, i] = 2
    return grid_in, grid_out


def task_edge_detection():
    """Highlight edges of filled region"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[1:5, 1:5] = 1

    grid_out = np.zeros((6, 6), dtype=int)
    grid_out[1:5, 1:5] = 1
    grid_out[2:4, 2:4] = 0  # Interior marked differently
    return grid_in, grid_out


def task_parity_grid():
    """XOR-based parity: output color based on parity of neighboring filled cells"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[1, 1] = 1
    grid_in[1, 3] = 1
    grid_in[3, 1] = 1
    grid_in[3, 3] = 1

    grid_out = np.zeros((5, 5), dtype=int)
    for i in range(5):
        for j in range(5):
            # Count filled neighbors
            neighbors = 0
            for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                ni, nj = i + di, j + dj
                if 0 <= ni < 5 and 0 <= nj < 5 and grid_in[ni, nj] > 0:
                    neighbors += 1
            grid_out[i, j] = neighbors % 2 + 1 if neighbors > 0 else 0

    return grid_in, grid_out


def task_distance_encoding():
    """Distance from source - diffusion-based"""
    grid_in = np.zeros((7, 7), dtype=int)
    grid_in[3, 3] = 1  # Source

    grid_out = np.zeros((7, 7), dtype=int)
    for i in range(7):
        for j in range(7):
            dist = abs(i - 3) + abs(j - 3)
            grid_out[i, j] = min(dist, 2)

    return grid_in, grid_out


def task_corner_marking():
    """Mark corners of rectangular region"""
    grid_in = np.zeros((7, 7), dtype=int)
    grid_in[2:5, 2:5] = 1

    grid_out = grid_in.copy()
    grid_out[2, 2] = 2
    grid_out[2, 4] = 2
    grid_out[4, 2] = 2
    grid_out[4, 4] = 2

    return grid_in, grid_out


def task_two_region_boundary():
    """Mark boundary between two regions"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[:3, :] = 1
    grid_in[3:, :] = 2

    grid_out = np.zeros((6, 6), dtype=int)
    grid_out[2, :] = 3
    grid_out[3, :] = 3

    return grid_in, grid_out


def task_line_extension():
    """Extend diagonal line"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[0, 0] = 1
    grid_in[1, 1] = 1
    grid_in[2, 2] = 1

    grid_out = np.zeros((6, 6), dtype=int)
    for i in range(6):
        grid_out[i, i] = 1

    return grid_in, grid_out


def task_inverse_region():
    """Invert all cells in a region"""
    grid_in = np.zeros((6, 6), dtype=int)
    grid_in[1:5, 1:5] = 1
    grid_in[2:4, 2:4] = 0

    grid_out = np.ones((6, 6), dtype=int)
    grid_out[1:5, 1:5] = 0
    grid_out[2:4, 2:4] = 1

    return grid_in, grid_out


def task_gradient_fade():
    """Create gradient from source"""
    grid_in = np.zeros((7, 7), dtype=int)
    grid_in[3, 3] = 1

    grid_out = np.zeros((7, 7), dtype=int)
    for i in range(7):
        for j in range(7):
            dist = max(abs(i - 3), abs(j - 3))
            grid_out[i, j] = max(0, 3 - dist)

    return grid_in, grid_out


def task_cross_pattern():
    """Extend cross/plus pattern"""
    grid_in = np.zeros((7, 7), dtype=int)
    grid_in[3, :] = 1
    grid_in[:, 3] = 1

    grid_out = grid_in.copy()

    return grid_in, grid_out


def task_rotational_symmetry():
    """Create 4-fold rotational symmetry"""
    grid_in = np.zeros((5, 5), dtype=int)
    grid_in[2, 1] = 1
    grid_in[2, 2] = 1

    grid_out = np.zeros((5, 5), dtype=int)
    grid_out[2, 1] = 1
    grid_out[2, 2] = 1
    grid_out[1, 2] = 1
    grid_out[2, 3] = 1
    grid_out[3, 2] = 1

    return grid_in, grid_out


# ============= MAIN EVALUATION =============

if __name__ == "__main__":
    results = []

    tasks = [
        ("Checkerboard Pattern", task_checkerboard(), 2, 8, 150),
        ("Diagonal Stripe", task_diagonal_stripe(), 3, 5, 120),
        ("Edge Detection", task_edge_detection(), 2, 10, 150),
        ("Parity Grid", task_parity_grid(), 3, 7, 150),
        ("Distance Encoding", task_distance_encoding(), 3, 12, 150),
        ("Corner Marking", task_corner_marking(), 3, 8, 150),
        ("Two Region Boundary", task_two_region_boundary(), 4, 6, 150),
        ("Line Extension", task_line_extension(), 2, 8, 150),
        ("Inverse Region", task_inverse_region(), 2, 8, 150),
        ("Gradient Fade", task_gradient_fade(), 4, 15, 150),
        ("Cross Pattern", task_cross_pattern(), 2, 5, 120),
        ("Rotational Symmetry", task_rotational_symmetry(), 2, 7, 150),
    ]

    for task_name, (grid_in, grid_out), n_colors, n_steps, iters in tasks:
        acc = run_task(grid_in, grid_out, task_name, n_colors, n_steps, iters)
        results.append((task_name, acc))

    # Summary
    print(f"\n\n{'='*70}")
    print("RESULTS SUMMARY")
    print(f"{'='*70}")
    for task_name, acc in sorted(results, key=lambda x: -x[1]):
        status = "✅" if acc >= 0.95 else "🟡" if acc >= 0.80 else "❌"
        print(f"{status} {task_name:30s} {acc:.2%}")
```
